# La adicción de los estudiantes a las redes sociales

## Introducción

El uso de las redes sociales (RRSS) en estudiantes impacta en el bienestar mental y el rendimiento académico. Este análisis exploratorio de datos (EDA) está motivado por la necesidad, encontrar en lo posible, la relación entre el uso de RRSS y el nivel de adicción percibido en estudiantes.

## Objetivo

Obtener insights a partir de un Análisis Exploratorio de Datos (EDA), utilizando visualizaciones para entender la estructura, características y relaciones en los datos sobre la adicción a redes sociales de los estudiantes.

# Columnas del dataset

**Student_ID/ID del estudiante**
Un identificador entero único asignado a cada encuestado para permitir la deduplicación y el seguimiento de registros individuales sin revelar información personal.

**Age/Edad**
La edad del estudiante en años completados al momento de la encuesta, utilizada para segmentar el análisis por grupo de edad y controlar las diferencias de desarrollo.

**Gender/Género**
El género autodeclarado por el estudiante, registrado como "Hombre" o "Mujer", para permitir desgloses demográficos en el uso y las mediciones de resultados.

**Academic_Level/Nivel académico**
El nivel educativo más alto en el que está matriculado actualmente el encuestado, con categorías: "Preparatoria", "Pregrado" o "Posgrado", lo que facilita análisis estratificados por etapa académica.

**Country/País**
El país de residencia donde el estudiante completó la encuesta, lo que permite realizar comparaciones entre países sobre el comportamiento y el impacto en las redes sociales.

**Avg_Daily_Usage_Hours/Promedio de horas de uso diario**
El promedio de horas diarias que el estudiante pasa en las redes sociales, calculado a partir de estimaciones de uso autodeclaradas entre semana y en fin de semana.

**Most_Used_Platform/Plataforma_Más_Usada**
La plataforma de redes sociales en la que el estudiante pasa más tiempo (p. ej., Instagram, Facebook, TikTok), utilizada para examinar los efectos específicos de cada plataforma.

**Affects_Academic_Performance/Afecta_Al_Rendimiento_Académico**
Un indicador binario (“Sí”/“No”) que refleja si el estudiante percibe que el uso de las redes sociales tiene un impacto negativo en su rendimiento académico.

**Sleep_Hours_Per_Night/Horas_De_Sueño_Por_Noche**
La duración promedio del sueño nocturno del encuestado en horas, proporcionada para investigar las correlaciones entre el tiempo frente a la pantalla y la calidad/cantidad del sueño.

**Mental_Health_Score/Puntuación_De_Salud_Mental**
Un número entero autoevaluado del 1 (malo) al 10 (excelente) que indica el bienestar mental general, lo que permite evaluar posibles asociaciones con los hábitos en redes sociales.

**Relationship_Status/Estado_Relacional**
El estado actual de la relación romántica del estudiante, categorizado como “Soltero”, “En_relación” o “Complicado”, para explorar el impacto de las redes sociales en la dinámica interpersonal.

**Conflicts_Over_Social_Media/Conflictos_en_Redes_Sociales**
El número de discusiones o desacuerdos que el estudiante informa haber tenido con familiares, amigos o parejas debido al uso de las redes sociales, lo que sirve como indicador de fricción social.

**Addicted_Score/Puntuación_de_Adicción**
Una puntuación compuesta de 1 (baja adicción) a 10 (alta adicción) basada en una escala de encuesta estandarizada (p. ej., la Escala de Adicción a las Redes Sociales de Bergen), que cuantifica el grado de uso problemático.

# Análisis Exploratorio Detallado (EDA) e Insights

**Pregunta 1:** Diferencia por Género y Nivel de Estudio

¿Existe una diferencia significativa en el nivel de adicción según el género o el nivel de estudio?

```python
# Análisis por Género
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='Addicted_Score', data=df, palette="YlGnBu")
plt.title('Distribución de Adicción por Género')
plt.show()

# Análisis por Nivel Académico
plt.figure(figsize=(10, 6))
sns.boxplot(x='Academic_Level', y='Addicted_Score', data=df, palette="YlGnBu")
plt.title('Distribución de Adicción por Nivel Académico')
plt.xlabel('Nivel Académico')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Distribución de Adicción por Género**

El insight del comparativo entre Hombres y Mujeres con la puntuación de Adicción sería que los estudianteshombes muestran una mediana de puntaje de adicción más alta que las estudiantes mujeres.

**Distribución de Adicción por Nivel Académico**

Los estudiantes de High School representan el grupo de mayor vulnerabilidad y consistencia en la alta adicción. Aunque para este grupo de riesgo se observa Outliers, que pueden presentar comportamientos o factores que hacen posean un nivel detectado de adicción bajo.

**Pregunta 2**: Relación entre Horas de Uso y Adicción

¿Cómo se relaciona el uso diario promedio (Avg_Daily_Usage_Hours) con el Addicted_Score?

```python
plt.figure(figsize=(10, 7))
sns.regplot(x='Avg_Daily_Usage_Hours', y='Addicted_Score', data=df, scatter_kws={'alpha':0.2}, line_kws={"color": "green"})
plt.title('Relación entre Horas de Uso Diario y Puntaje de Adicción')
plt.xlabel('Horas de Uso Diario Promedio')
plt.ylabel('Puntaje de Adicción')
plt.show()
```

**Relación entre Horas de Uso Diario y Puntaje de Adicción**

El uso de más horas está directamente asociado con una mayor puntuación de adicción.

Correlación entre variables.

```python
# Columnas numéricas
cols = [
    'Age',
    'Avg_Daily_Usage_Hours',
    'Sleep_Hours_Per_Night',
    'Mental_Health_Score',
    'Conflicts_Over_Social_Media',
    'Addicted_Score'
]
df_numeric = df[cols]

#  matriz de correlación
corr_matrix = df_numeric.corr()

# heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu', fmt=".2f", linewidths=0.5)
plt.title("Mapa de calor de correlaciones entre variables")
plt.show()
```

**Insights encontrados.**
- Los estudiantes más jóvenes tienden a tener mayor adicción a redes sociales.
- A mayor tiempo diario en redes, mayor nivel de adicción.
- La adicción se asocia con menos horas de sueño y peor salud mental.
- Los conflictos sociales aumentan significativamente con el nivel de adicción.


Pregunta 3 : Conflictos sociales por nivel de adicción

¿A mayor nivel de adicción mayor nivel de conflicto social?

```python
# Crear categorías de adicción
df['Adiction_Level'] = pd.cut(df['Addicted_Score'],
                              bins=[0, 3, 6, 10],
                              labels=['Bajo', 'Medio', 'Alto'])

# Promedio de conflictos por nivel
conflict_avg = df.groupby('Adiction_Level')['Conflicts_Over_Social_Media'].mean().reset_index()

# Gráfico de barras
plt.figure(figsize=(8, 6))
sns.barplot(x='Adiction_Level', y='Conflicts_Over_Social_Media', data=conflict_avg, palette='YlGnBu')
plt.title('Promedio de conflictos sociales por nivel de adicción')
plt.xlabel('Nivel de adicción')
plt.ylabel('Conflictos sociales promedio')
plt.show()
```
